```python
import openai
from google.colab import userdata
APIKEY=userdata.get('OPENAI_API_KEY')
```

```python
from openai import OpenAI

# Initialize the OpenAI client
client = OpenAI(
    api_key=APIKEY  # You can also set this manually, but using an environment variable is safer
)
```

```python
test = Label.iloc[2000:3000]
print(test)
for i in test:
  print(i)
```

```python
responses = []

# Loop through each input and call the API
for user_input in description:
    # Send the API request
  response = client.chat.completions.create(
        model="ft:gpt-4o-2024-08-06:personal:tabularllm:AWDWklhl",
        messages=[
            {"role": "user", "content": [{"type": "text", "text": user_input}]}
        ],
        response_format={"type": "text"},
        temperature=0.8,
        max_tokens=3,
        top_p=1,
        frequency_penalty=0.4,
        presence_penalty=0
  )

  response = response.to_dict()# Extract the assistant's response
  assistant_response = response['choices'][0]['message']['content']
  # Print the assistant's response
  print(assistant_response)
    # Store or print the response
  responses.append(assistant_response)


# Optionally, you can save the responses to a file or handle them further
with open("model_responses.txt", "w") as f:
    for response in enumerate(responses):
        f.write(f"{response}\n")
```

```python
accuracies2.clear()
```

```python
accuracy = 0
total=0
with open("model_responses.txt", "r") as f:
    for line, marital_status in zip(f, test):
        # Clean model response
        model_responses = line.strip().split(',')
        model_responses = model_responses[1].split(')')[0]
        model_responses = model_responses.lower()
        model_responses = model_responses.replace("'", "")
        model_responses = model_responses.replace("the marital status of the person is", "").strip()

        # Clean marital status
        total+=1
        print(marital_status, model_responses)
        marital_status = marital_status.lower().strip()



        if model_responses == marital_status:
            accuracy += 1
print(accuracy, total)
accuracy = (accuracy / total) * 100
accuracies2.append(accuracy)
print(f"Final accuracy: {accuracy}")
```

```python
import matplotlib.pyplot as plt

# Example: You already have a list of accuracies for multiple models
# Replace this list with the actual accuracy of your models
  # List of accuracies for models (in %)

# Model numbers (just indices, e.g., Model 1, Model 2, etc.)

model_numbers = range(1, len(accuracies2) + 1)
model_checkpoints=["1500checkpoint" ,"1000checkpoint" ,"500checkpoint"]
colorrepresent = ["red", "blue" , "green"]
# Plot the vertical bar graph
plt.figure(figsize=(10, 6))
plt.bar(model_numbers, accuracies2, color=[colorrepresent[i] for i in range(len(model_numbers))]) # Assumed colorrepresent is an iterable with sufficient elements

# Customize the graph
plt.xlabel('Model Number')
plt.ylabel('Accuracy (%)')
plt.title('Model Accuracy Comparison')
plt.ylim(0, 100)  # Y-axis from 0 to 100% accuracy
plt.xticks(model_numbers)  # Set x-ticks as model numbers
plt.grid(True, axis='y', linestyle='--', alpha=0.7)  # Horizontal grid lines
plt.show()
```
